### <span style="color:darkviolet"><left>Correcting Spelling Mistakes

#### 1.) Bangalore to Bengaluru

```python
# Let us correct Bangalore to Bengaluru
def city_name(name):
    if name == 'Bangalore':
        return 'Bengaluru'
    else:
        return name

df1.city = df1.city.apply(city_name)
df1.city.unique()
```

#### 2.) Rising Pune Supergiant to Rising Pune Supergiants

```python
# Correcting 'Rising Pune Supergiant' to 'Rising Pune Supergiants' where ever this misatke is present
def team_name(name):
    if name == 'Rising Pune Supergiant':
        return 'Rising Pune Supergiants'
    else:
        return name

df1.team1, df1.team2, df1.toss_winner, df1.winner,df2.bowling_team,df2.batting_team = df1.team1.apply(team_name), df1.team2.apply(team_name), df1.toss_winner.apply(team_name), df1.winner.apply(team_name),df2.bowling_team.apply(team_name),df2.batting_team.apply(team_name)
print(df1.team1.unique())# Check minimum in one if the function has been applied
```

```python
df1.replace({'M. Chinnaswamy Stadium':'M Chinnaswamy Stadium'},inplace=True)
```

#### 4.) M. A. Chidambaram Stadium to MA Chidambaram Stadium, Chepauk

```python
df1.replace({'M. A. Chidambaram Stadium':'MA Chidambaram Stadium, Chepauk'},inplace=True)
```
